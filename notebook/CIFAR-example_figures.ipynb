```python
# Import original train images
imgs = [Image.open(path) for path in sorted(glob(figure_dir + 'train_labeled*.png'))]

# Interpolation
gamma = 0.65
pairs = [(0,1), (1,2), (2,3), (3,0)]
imgs_array = [np.array(img) for img in imgs]
imgs_mixup = [gamma*imgs_array[pair[0]] + (1-gamma)*imgs_array[pair[1]] for pair in pairs]
imgs = [Image.fromarray(img.astype('uint8')) for img in imgs_mixup]

# 2 by 2 subplots
f, ax = plt.subplots(2,2)
ax[0,0].imshow(imgs[0])
ax[0,0].axis('off')
ax[0,1].imshow(imgs[1])
ax[0,1].axis('off')
ax[1,0].imshow(imgs[2])
ax[1,0].axis('off')
ax[1,1].imshow(imgs[3])
ax[1,1].axis('off')
plt.subplots_adjust(wspace=-0.5, hspace=0.1)
#plt.savefig(save_dir + 'mixup.pdf')
plt.show()
```
