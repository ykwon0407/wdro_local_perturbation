```python
from glob import glob
import json
import numpy as np
from scipy.stats import ttest_ind
```

```python
for p in [0.01,0.02,0.04]:
    test_acc = {}
    for model in ['erm', 'erm_0.004', 'mixup','l2_0.004']:
        noise_path_list = sorted(glob('../experiments/{}/cifar10.*@50000-1/noise.txt'.format(model)))
        acc_path_list = sorted(glob('../experiments/{}/cifar10.*@50000-1/accuracies.txt'.format(model)))
        tmp = []
        tmp_acc = []
        for path in noise_path_list:
            f = json.load(open(path, "r"))
            noise_setting = {'Mode': 's&p', 'p': p, 'seed': None, 'steps': '06553600'}
            noise_dict = f[str(noise_setting)]
            tmp.append(noise_dict['noise_acc'][2])
        for path in acc_path_list:
            f = json.load(open(path, "r"))
            tmp_acc.append(f['epoch99'][2])
            
        test_acc.update({model:np.array(tmp)-np.array(tmp_acc)})
    best_model = list(test_acc.keys())[np.argmax([np.mean(test_acc[model]) for model in test_acc.keys()])]
    #print(best_model)
    for model in test_acc.keys():
        pvalue = ttest_ind(test_acc[best_model], test_acc[model], equal_var=False).pvalue
        print('model=' + model + ' p = %.2f  mean = %.1f  std=%.1f  p-value=%.3f is equivalent?=%r' 
              % (p,np.mean(test_acc[model]),np.std(test_acc[model]), pvalue, pvalue>=0.05))
```

```python
for p in [0.01,0.02,0.04]:
    test_acc = {}
    for model in ['erm', 'mixup', 'erm_0.004', 'l2_0.004']:
        noise_path_list = glob('../experiments/{}/cifar100.*@50000-1/noise.txt'.format(model))
        tmp = []
        for path in noise_path_list:
            f = json.load(open(path, "r"))
            noise_setting = {'Mode': 's&p', 'p': p, 'seed': None, 'steps': '06553600'}
            noise_dict = f[str(noise_setting)]
            tmp.append(noise_dict['noise_acc'][2])
        test_acc.update({model:tmp})
    best_model = list(test_acc.keys())[np.argmax([np.mean(test_acc[model]) for model in test_acc.keys()])]
    #print(best_model)
    for model in test_acc.keys():
        pvalue = ttest_ind(test_acc[best_model], test_acc[model], equal_var=False).pvalue
        print('model=' + model + ' p = %.2f  mean = %.1f  std=%.1f  p-value=%.3f is equivalent?=%r' 
              % (p,np.mean(test_acc[model]),np.std(test_acc[model]), pvalue, pvalue>=0.05))
```

```python
for model in ['erm', 'l2']:
    test_acc = {}
    for lam in ['', '_0.004', '_0.016', '_0.064']:
        if model+lam == 'l2': 
            noise_path_list = glob('../experiments/mixup/cifar100.*@50000-1/noise.txt')
        else:
            noise_path_list = glob('../experiments/{}/cifar100.*@50000-1/noise.txt'.format(model+lam))
        tmp = []
        for path in noise_path_list:
            f = json.load(open(path, "r"))
            noise_setting = {'Mode': 's&p', 'p': 0.01, 'seed': None, 'steps': '06553600'}
            noise_dict = f[str(noise_setting)]
            tmp.append(noise_dict['noise_acc'][2])
        test_acc.update({model+lam:tmp})
    best_model = list(test_acc.keys())[np.argmax([np.mean(test_acc[model]) for model in test_acc.keys()])]
    #print(best_model)
    #print(test_acc)
    for model in test_acc.keys():
        pvalue = ttest_ind(test_acc[best_model], test_acc[model], equal_var=False).pvalue
        print('model=' + model + ' mean = %.1f  std=%.1f  p-value=%.3f is equivalent?=%r' 
              % (np.mean(test_acc[model]),np.std(test_acc[model]), pvalue, pvalue>=0.05))
```
